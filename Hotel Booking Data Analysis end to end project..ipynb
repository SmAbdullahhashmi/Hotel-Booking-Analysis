# Data Analysis and Visualizations

```python
cancelled_perc = htl['is_canceled'].value_counts(normalize = True)
print(cancelled_perc)
plt.figure(figsize = (5,4))
plt.title('Reservation status count')
plt.bar(['Not canceled','Canceled'],htl['is_canceled'].value_counts(), edgecolor = 'k', width = 0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 4))
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=htl, palette='Blues')
legend_labels, _ = ax1.get_legend_handles_labels()

# Position the legend outside the plot area
ax1.legend(legend_labels, ['Not Canceled', 'Canceled'], title='Reservation Status', loc='upper right', bbox_to_anchor=(1.2, 1))

plt.title('Reservation Status in Different Hotels', size=20)
plt.xlabel('Hotel')
plt.ylabel('Number of Reservations')
plt.show()
```

```python
resort_hotel = htl[htl['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = htl[htl['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 30)
plt.plot(resort_hotel.index,resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

```python
htl['month'] =htl['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x = 'month',hue = 'is_canceled',data = htl,palette = 'bright')
legend_labels,_ = ax1. get_legend_handles_labels()
plt.title('Reservation status per month',size = 20)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['Not Cancelled','Cancelled'])
plt.show()
```

```python
plt.figure(figsize = (15, 8))
plt.title('ADR per month', fontsize=30)
sns.barplot(x='month', y='adr', data= htl[htl['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

```python
cancelled_data = htl[htl['is_canceled'] == 1]
top_10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize = (8,8))
plt.title('Top 10 countries with reservation canceled')
plt.pie(top_10_country,autopct = '%.2f',labels = top_10_country.index)
plt.show()
```

```python
htl['market_segment'].value_counts()
```

```python
htl['market_segment'].value_counts(normalize = True)
```

```python
cancelled_data['market_segment'].value_counts(normalize = True)
```

```python
cancelled_htl_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_htl_adr.reset_index(inplace = True)
cancelled_htl_adr.sort_values('reservation_status_date', inplace = True)

not_cancelled_data = htl[htl['is_canceled'] == 0]
not_cancelled_htl_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_htl_adr.reset_index(inplace = True)
not_cancelled_htl_adr.sort_values('reservation_status_date', inplace = True)


plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_htl_adr['reservation_status_date'], not_cancelled_htl_adr['adr'], label='Not Cancelled')
plt.plot(cancelled_htl_adr['reservation_status_date'], cancelled_htl_adr['adr'], label='Cancelled')
plt.legend()
plt.show()
```

```python
cancelled_htl_adr = cancelled_htl_adr[(cancelled_htl_adr['reservation_status_date']>'2016')& (cancelled_htl_adr['reservation_status_date']<'2017-09')]
not_cancelled_htl_adr = not_cancelled_htl_adr[(not_cancelled_htl_adr['reservation_status_date']>'2016')& (not_cancelled_htl_adr['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate', fontsize = 30)
plt.plot(not_cancelled_htl_adr['reservation_status_date'], not_cancelled_htl_adr['adr'], label='Not Cancelled')
plt.plot(cancelled_htl_adr['reservation_status_date'], cancelled_htl_adr['adr'], label='Cancelled')
plt.legend(fontsize = 20)
plt.show()
```

# Suggestions

1. Cancellation rates rise as the price does. In order to prevent cancellation of reservations,
   hotels could work on their pricing strategies and try to lower the rates for specific hotels based on locations.
   They can also provide some discounts to the consumers.

2. As the ratio of the cancellation and not cancellation of the resort hotel is higher in the resort hotel than the city hotels.
   so the hotels should provide a reasonable discount on the room prices on weekends or on holidays.
    
3. In the month of January, hotels can start campaigns or marketing with a reasonable amount to increase their revenue as the 
   cancellation is the highest in thist month.
    
4. They can also increase the quality of their hotels and their services mainly in 
   Portugal to reduce the cancellation Rate.
